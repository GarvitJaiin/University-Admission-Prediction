```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\User\Downloads\Admission_Predict.csv')
df
```

```python
df.info()
```

```python
df.drop('Serial No.',axis =1 , inplace =True)
df.head()
```

```python
df.describe()
```

```python
df['Admission'] = [ 1 if x >= 0.785 else 0 for x in df['Chance of Admit ']]
df.head()
```

```python
df.corr()
```

```python
df.drop(['Chance of Admit '],axis =1, inplace =True)
```

```python
X = df.loc[:,df.columns != 'Admission']
Y = df['Admission']
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
X
```

```python
Y = Y.values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state =0)
```

```python
from sklearn.linear_model import LogisticRegression
Lr = LogisticRegression().fit(X_train,Y_train)
```

```python
y_hat = Lr.predict(X_test)
```

```python
y_hat
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
score = accuracy_score(Y_test,y_hat)
score
```

```python
y_hat_probability = Lr.predict_proba(X_test)
y_hat_probability
```

```python
print(classification_report(Y_test,y_hat))
```

```python
confusion_matrix(Y_test,y_hat,labels=[1,0])
```
